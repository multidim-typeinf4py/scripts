```python
import pandas as pd
import pathlib
from scripts.infer.structure import CrossDomainTypes4Py

dataset = CrossDomainTypes4Py(pathlib.Path("/nfs/data/students/bsparks/mdti4py-dataset-pool/cdt4py"))
troved = pd.read_csv(dataset.dataset_root / "troved.csv")
```

```python
validation_author_repos = [
    str(dataset.author_repo(project))
    for project in dataset.el_validation_set()
]
display(troved.repository.value_counts().index.isin(validation_author_repos).sum())
validation_troves = troved[troved.repository.isin(validation_author_repos)]
display(validation_troves.classifier.value_counts())
```

```python
test_author_repos = [
    str(dataset.author_repo(project))
    for project in dataset.el_test_set()
]
display(troved.repository.value_counts().index.isin(test_author_repos).sum())
test_troves = troved[troved.repository.isin(test_author_repos)]
display(test_troves.classifier.value_counts())
```

```python
exploded_troved = troved.classifier.str.split(" :: ", expand=True)
troved = troved.drop(columns=["classifier"])
troved[[1, 2, 3, 4]] = exploded_troved[[1, 2, 3, 4]]
display(troved)
```

```python
final_troved = troved[["repository", 1, 2, 3]]
display(final_troved)
```

```python
# Outermost is irrelevant for us for now
final_troved = final_troved.drop(columns=[3])
display(final_troved)
```

```python
final_troved[[1, 2]].value_counts().sort_index()
```

```python
# Artificial Intelligence and Artifical Life -> AI
ai_mask = (final_troved[1] == "Scientific/Engineering") & (final_troved[2].isin(["Artificial Intelligence", "Artificial Life"]))
final_troved.loc[ai_mask, 2] = "AI"

# Bio-Informatics, Chemistry, Physics, Medical Science, Mathematics -> Science
research_mask = (final_troved[1] == "Scientific/Engineering") & (final_troved[2].isin(
    ["Atmospheric Science", "Bio-Informatics", "Chemistry", "Physics", "Medical Science Apps.", "Mathematics"]
))
final_troved.loc[research_mask, 2] = "Science"

# Information Analysis, Visualization, GIS -> Data Science
ds_mask = (final_troved[1] == "Scientific/Engineering") & (final_troved[2].isin(
    ["Information Analysis", "Visualization", "GIS"]
))
final_troved.loc[ds_mask, 2] = "Data Science"

# Human Machine Interfaces, Interface Engine/Protocol Translator -> Interface System
interface_mask = (final_troved[1] == "Scientific/Engineering") & (final_troved[2].isin(
    ["Human Machine Interfaces", "Interface Engine/Protocol Translator"]
))
final_troved.loc[interface_mask, 2] = "Interface System"
```
